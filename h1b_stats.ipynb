## Inferential Statistics

Null hypothesis: This feature (employer name, occupation, state, etc.) is not related to certified H1B cases.

Alternate hypothesis: This feature and certified H1B cases are related to one another.

Methods: Pearson's Chi-square test

```python
#Import necessary packages
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("h1b_kaggle_clean.csv",index_col=0)
data=data[data['CASE_STATUS'].isin(['CERTIFIED','DENIED'])]
data.head()
```

```python
# Crosstab using full time position and case status
cont_table = pd.crosstab(data.FULL_TIME_POSITION, data.CASE_STATUS)
cont_table
```

```python
# Compute Chi Square and get p value
chi2, p, dof, ex = stats.chi2_contingency(cont_table)

print("chi2 =", chi2)
print("p =",p)

# 95% confidence interval
chi2_crit = stats.chi2.ppf(q = 0.95, df = dof)
print("chi2_crit =", chi2_crit)
```

```python
# Bin wages
#Binning:
def binning(col, cut_points, labels=None):
  #Define min and max values:
  minval = col.min()
  maxval = col.max()

  #create list by adding min and max to cut_points
  break_points = [minval] + cut_points + [maxval]

  #if no labels provided, use default labels 0 ... (n-1)
  if not labels:
    labels = range(len(cut_points)+1)

  #Binning using cut function of pandas
  colBin = pd.cut(col,bins=break_points,labels=labels,include_lowest=True)
  return colBin

#Binning age:
summary = np.percentile(data.PREVAILING_WAGE, [33, 67])
# print(summary) - find cut off points

cut_points = [58365,74984]
labels = ["low","medium","high"]
data["WageBin"] = binning(data["PREVAILING_WAGE"], cut_points, labels)
print(pd.value_counts(data["WageBin"], sort=False))
data.head()
```

```python
# Crosstab wagebin and case status
cont_table6 = pd.crosstab(data.WageBin, data.CASE_STATUS)
cont_table6
```

```python
# Compute Chi Square and get p value
chi2, p, dof, ex = stats.chi2_contingency(cont_table6)

print("chi2 =", chi2)
print("p =",p)

chi2_crit = stats.chi2.ppf(q = 0.95, df = dof)
print("chi2_crit =", chi2_crit)
```
